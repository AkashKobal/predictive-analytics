```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("car_data.csv")
```

```python
display(df.head(5))
```

```python
X = df.iloc[:400,[2,3]].values
y = df.iloc[:400,-1].values
print(X)
```

```python
sampled_df = df.sample(n=400) # selecting the random 400 rows
X = sampled_df.iloc[:,[2,3]].values
y = sampled_df.iloc[:,-1].values
print(X)
```

```python
#splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling
#Import the necessary libraries:
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

#create a logistic regression classifier
classifier = LogisticRegression()
```

```python
#create a standardscaler object and fit it to the training data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)

# scale the feature in the test set(X_train):
X_test = sc.transform(X_test)

# if y_train is a 2D array, you need to flatten it to a 1D array using ravel() function:
y_train = y_train.ravel()

#Fit the logistic regression model using the scaled training data:
classifier.fit(X_train, y_train)
print(X_train)
```

```python
print(y_train)
# 0 means not buy, 1 means buy
# 1D array
```

```python
print(classifier.predict(sc.transform([[49,74000]])))
```

```python
#Making a prediction o the test set
y_pred = classifier.predict(X_test)

#Concatenating the predicted and actual values in a single column:
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Importing necessary library for evaluation
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

#Generating the confusion matrix
cm = confusion_matrix(y_test, y_pred)

#plotting the confusion matrix using seaborn heatmap
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
#Calculating and printing the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
print(f'Confusion Matrix:\n{cm}')
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
#import the necessary libraries
from matplotlib.colors import ListedColormap
#Transform the scaled training feature back to the original scale using the inverse_transform() function:
X_set, y_set = sc.inverse_transform(X_train), y_train

#Create a meshgridp to plot the decision boundaries
X1, X2 = np.meshgrid(np.arange(start = X_set[:,0].min()-10, stop = X_set[:,0].max()+10, step = 0.25),
                     np.arange(start = X_set[:,1].min()-1000, stop = X_set[:,1].max()+1000, step = 0.25))
                     
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('slategrey','slateblue')))
#set the plot limits:
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Scatter plot the original data points with different colors for each class:
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('slategrey','slateblue'))(i), label = j)
# set plot title, x-axis label, y-axis label, and legend:
plt.title('Random Forest Classification(Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()

#Display the plot:
plt.show()
```

```python
#Probality
#create a new data point to predict
new_data = [[35,50000]]
#acale the new data using the same scaler object used to scale the training data
new_data_scaled = sc.transform(new_data)
#predict the new data using the classifier
prediction = classifier.predict(new_data_scaled)
print(prediction)
```

```python
#New prediction for 6(potential) new customers:
new_data_2 = [[35, 50000], [42, 15700], [51, 65000], [23, 45600], [36, 42000], [41, 72000]]
df = pd.DataFrame(new_data_2, columns = ['Age', 'Annual Salary'])
df.index = df.index + 1
display(df)
```

```python
from tabulate import tabulate
results = []
for i, data in enumerate(new_data_2):
    #scale the new data using the same scaler object used to scale the training data
    data_scaled = sc.transform([data])
    # make a prediction using the trained classifier
    prediction = classifier.predict(data_scaled)
    # add the result to the table
    if prediction ==1:
        prediction_text = "Yes"
    else:
        prediction_text = "No"
    results.append([i+1, data[0], data[1], prediction_text])
    #print the table
    headers = ["Customer", "Age", "Annual Salary","prediction"]
    print(tabulate(results, headers=headers))
```
