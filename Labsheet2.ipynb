```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv("car_data.csv")
```

```python
display(df.head(5))
```

```python
X = df.iloc[:400,[2,3]].values
y = df.iloc[:400,-1].values
print(X)
```

```python
sampled_df = df.sample(n=400) # selecting the random 400 rows
X = sampled_df.iloc[:,[2,3]].values
y = sampled_df.iloc[:,-1].values
print(X)
```

```python
#splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling
#Import the necessary libraries:
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

#create a logistic regression classifier
classifier = LogisticRegression()
```

```python
#create a standardscaler object and fit it to the training data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)

# scale the feature in the test set(X_train):
X_test = sc.transform(X_test)

# if y_train is a 2D array, you need to flatten it to a 1D array using ravel() function:
y_train = y_train.ravel()

#Fit the logistic regression model using the scaled training data:
classifier.fit(X_train, y_train)
print(X_train)
```

```python
print(y_train)
# 0 means not buy, 1 means buy
# 1D array
```

```python
print(classifier.predict(sc.transform([[49,74000]])))
```

```python
#Making a prediction o the test set
y_pred = classifier.predict(X_test)

#Concatenating the predicted and actual values in a single column:
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Importing necessary library for evaluation
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

#Generating the confusion matrix
cm = confusion_matrix(y_test, y_pred)

#plotting the confusion matrix using seaborn heatmap
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
#Calculating and printing the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
print(f'Confusion Matrix:\n{cm}')
print(f'Accuracy: {accuracy*100:.2f}%')
```
